## Using linear regression to predict future average housing prices

```python
# Load housing data again
housing = pd.read_csv(r"C:\Users\suhai\Downloads\price_paid_records.csv")
```

```python
housing['Date of Transfer'] = pd.to_datetime(housing['Date of Transfer'])
```

```python
# format date column
date = "Date of Transfer"
housing[date] = pd.to_datetime(housing[date], format='%d/%m/%y')

# Set the date as the index
housing.set_index(date, inplace=True)

# Group by year and month, then calculate the mean of price
monthly = housing.resample('M')['Price'].mean().reset_index()

# Display the result
print(monthly)
```

```python
monthly.keys()
```

```python
# plot the graph of housing price against year-month
plt.plot(monthly['Date of Transfer'], monthly['Price'],'tab:blue')
plt.xlabel('year')
plt.ylabel('Average UK housing price ()')
plt.title('Average housing price against year', fontsize = 11)
plt.grid(True,linestyle='--', alpha=0.6)
```

```python
print(monthly.head(5))
```

```python
# Create a feature 'Previous Close'
monthly['Prev Price'] = monthly['Price'].shift(1)

# Drop any rows with NaN values
monthly.dropna(inplace=True)
```

```python
# Define features (X) and target (y)
# natural log the feature and target
x = np.log10(monthly[['Prev Price']])
y = np.log10(monthly['Price'])
```

```python
# 80% train and 20% test
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, shuffle=False)
```

```python
# Initialize the model
model = LinearRegression()

# Train the model
model.fit(X_train, y_train)
```

```python
# Get the dates so you can plot a graph
pre = monthly.iloc[215:, :] # 0.2 is 20 % of the rows
#print(pre)
```

```python
# Predict on the test set
y_pred = model.predict(X_test)

# Plot the results
# Result show the last 20% of the data. This is the test of the model.
plt.figure(figsize=(10, 6))
plt.plot(pre['Date of Transfer'], y_test, label='Actual Prices')
plt.plot(pre['Date of Transfer'], y_pred, label='Predicted Prices')
plt.ylabel(r'$log_{10}$(price) ()')
plt.xlabel('year-month')
plt.legend()
plt.title('Actual vs Predicted Prices')
plt.show()

# Calculate the mean squared error
mse = np.mean((y_test - y_pred) ** 2)
print(f'Mean Squared Error: {mse.round(5)}')
```

```python
# Predict future price
last_known_price = np.log10(monthly['Price'].iloc[-1])
next_price = model.predict([[last_known_price]])
print(f'Next predicted price: {10**next_price[0]}')
```

```python
# Predict price for the next 10 years
future_price = np.zeros(120)
last_price = y_pred[-1]
for i in range(120):
    future = model.predict([[last_price]])
    future_price[i] = future
    last_price = future[0]
```

```python
#print(future_price)
```

```python
plt.figure(figsize=(10, 6))
future_dates = pd.date_range("2017-06-30", periods= 120, freq='M')
plt.plot(pre['Date of Transfer'], y_test, label='Actual Prices')
plt.plot(pre['Date of Transfer'], y_pred, label='Predicted Prices')
plt.plot(future_dates, future_price, '--', color = 'orange', label = 'Future prices')
plt.ylabel('$log_{10}$(price) ()')
plt.xlabel('year')
plt.legend()
plt.title('Predicting future prices')
plt.tight_layout()
```
