### plotting noise against frequencies

```python
def RootMeanSquare(image):
    rms = np.sqrt(np.abs(np.mean(image)))
    return rms
def newtime(new_number):
    old_number = 1000
    time = np.sqrt(float(old_number)/(float(new_number)))
    return time

#change the number of hours the signal is measured
new_number = 10.0
Noise_100 = np.zeros((128,128,342))
rmsNoise_100 = np.zeros(342)
meanNoise_100 = np.zeros(342)
for i in range(342):
    noise_100 = noise[:,:,i]*newtime(new_number)
    rmsNoise_100[i] = RootMeanSquare(noise_100)
    meanNoise_100[i] = np.mean(noise_100)
    Noise_100[:,:,i] = noise_100[:,:]

rmsNoise_1000 = np.zeros(342)
meanNoise_1000 = np.zeros(342)
#rms of the 1000 hours noise
for i in range(342):
    rmsNoise_1000[i] = RootMeanSquare(noise[:,:,i])
    meanNoise_1000[i] = np.mean(noise[:,:,i])

new_number = 10000
Noise_10000 = np.zeros((128,128,342))
rmsNoise_10000 = np.zeros(342)
meanNoise_10000 = np.zeros(342)
for i in range(342):
    noise_10000 = noise[:,:,i]*newtime(new_number)
    rmsNoise_10000[i]  = RootMeanSquare(noise_10000)
    meanNoise_10000[i] = np.mean(noise_10000)
    Noise_10000[:,:,i] = noise_10000[:,:]
```

```python
#plot figures    
plt.rcParams.update({'font.size' : 8})
x = np.linspace(108,208.3,342)
plt.figure(figsize = (6,3))

plt.plot(x,rmsNoise_100,'r-',label='     100 hours')
plt.plot(x,rmsNoise_1000,'g-',label='  1,000 hours')
plt.plot(x,rmsNoise_10000,'b-',label='10,000 hours')
plt.xlabel('Frequency (MHz)')
plt.ylabel('RMS signal (k)')
#plt.title('RMS against frequency', fontsize = 10)
plt.legend()
plt.grid(True, linestyle= '--', alpha = 0.5)
plt.tight_layout()
#plt.savefig( folder +'RMS_noise.png')
```

```python
#plot figures    
x = np.linspace(108,208.3,342)
fig, ax = plt.subplots(1, 2,figsize=(10,4))

ax[0].plot(x,rmsNoise_100,'r-',label='100 hours')
ax[0].plot(x,rmsNoise_1000,'b-',label='1000 hours')
ax[0].plot(x,rmsNoise_10000,'g-',label='1000 hours')
ax[0].set_xlabel('Frequency (MHz)')
ax[0].set_ylabel('rms signal (k)')
ax[0].set_title('rms against frequency')
ax[0].legend()

ax[1].plot(x,meanNoise_100,'r-',label='100 hours')
ax[1].plot(x,meanNoise_1000,'b-',label='1000 hours')
ax[1].plot(x,meanNoise_10000,'g-',label='10000 hours')
ax[1].set_xlabel('Frequency (MHz)')
ax[1].set_ylabel('mean signal (k)')
ax[1].set_title('mean against frequency')
ax[1].legend()

plt.tight_layout()

#plt.savefig('images/mean_and_rms.png')

#save different hours files
#hdu = fits.PrimaryHDU(Noise_10000)
#hdul = fits.HDUList([hdu])
#hdul.writeto('fits_files/Noise_10000.fits',overwrite=True)

#hdu = fits.PrimaryHDU(Noise_100)
#hdul = fits.HDUList([hdu])
#hdul.writeto('fits_files/Noise_100.fits',overwrite=True)
```

### Plotting foregrounds, 21cm and noise against frequency

```python
#load foregound separately
indata_cs = fits.getdata('boxes/test.fits').T
indata_cs = indata_cs/1000   #convert to mili kelvins
print(indata_cs.shape)
```

```python
#load foregound separately
light = fits.getdata('boxes/test.fits').T
light = indata_cs/1000   #convert to mili kelvins

var_gal = np.zeros(342)
var_syn = np.zeros(342)
var_free = np.zeros(342)
var_noise = np.zeros(342)
var_light = np.zeros(342)
for i in range(342):
    var_syn[i] = np.var(syn_test[:,:,i])
    var_gal[i] = np.var(egfg_max[:,:,i]) #change to max exgal
    var_free[i] = np.var(ff_test[:,:,i])
    var_noise[i] = np.var(noise[:,:,i])
    var_light[i] = np.var(light[:,:,i])

x = np.linspace(108,208.3,342)
fig = plt.figure(figsize=(10,10))

ax = fig.add_subplot(2,2,1)
ax.plot(x,var_gal,'r-',label = 'extra galatic')
ax.plot(x,var_syn,'b-', label = 'syncrotron emission')
ax.plot(x,var_free*10000,'g-', label='free free emission')
ax.plot(x, var_noise*5e4,'tab:purple', label = 'noise')
ax.plot(x, var_light*1e11, 'c-', label = '21cm signal')
ax.set_xlabel('frequency (MHZ)')
ax.set_ylabel('variance (K)')
ax.set_title('variance against frequency')
ax.legend()

ax = fig.add_subplot(2,2,2)
ax.plot(x,var_free,'g-', label='free free emission')
ax.set_xlabel('frequency (MHZ)')
ax.set_ylabel('variance (K)')
ax.set_title('variance against frequency')
ax.legend()

ax = fig.add_subplot(2,2,3)
ax.plot(x, var_noise,'tab:purple', label = 'noise')
ax.set_xlabel('frequency (MHZ)')
ax.set_ylabel('variance (K)')
ax.set_title('variance against frequency')
ax.legend()

ax = fig.add_subplot(2,2,4)
ax.plot(x, var_light, 'c-', label = '21cm signal')
ax.set_xlabel('frequency (MHZ)')
ax.set_ylabel('variance (K)')
ax.set_title('variance against frequency')
ax.legend()

plt.tight_layout()

#plt.savefig('images/new_variance_against_frequency.png')
```

```python
# plot of all foreground, noise and 21cm signal

x = np.linspace(108,208.3,342)
plt.figure(figsize=(4,5))
plt.rcParams.update({'font.size' : 9})

plt.plot(x,var_syn, color ='#ff7f0e', label = 'Synchrotron emission')
plt.plot(x,var_gal, color ='#1f77b4',label = 'Extra galactic emission')
plt.plot(x,var_free*5e3, color ='#2ca02c', label= r'Free-free emission: $5 \times 10^3$')
plt.plot(x, var_noise*3e4, color ='#d62728', label = r'Thermal noise: $3 \times 10^4$')
plt.plot(x, var_light*2e11, color ='black', label = r'Cosmological signal: $2 \times 10^{11}$')
plt.xlabel('Frequency (MHZ)')
plt.ylabel('Variance (K)')
#plt.title('variance against frequency')
plt.legend()
plt.grid(True, linestyle= '--', alpha = 0.5)
plt.tight_layout()
plt.savefig( folder +'AllParameters.png')
```
