```python
import numpy as np
import sys
import matplotlib.pyplot as plt
from scipy import signal
from astropy.io import fits
from sklearn.decomposition import FastICA, PCA
from scipy.stats import pearsonr as pear
from PIL import Image, ImageDraw
import matplotlib.image as mpimg
```

```python
#load images
image1 = mpimg.imread('images/Variance_noise100.png')
image2 = mpimg.imread('images/Variance_noise1000.png')
image3 = mpimg.imread('images/Variance_noise10000.png')
```

```python
fig, axes = plt.subplots(3,1,figsize = (20,20))

#display images

axes[0].imshow(image1)
axes[0].set_title('100 hours observation')

axes[1].imshow(image2)
axes[1].set_title('1000 hours observation')

axes[2].imshow(image3)
axes[2].set_title('10000 hours observation')

for ax in axes:
    ax.axis('off')
    
plt.tight_layout()
plt.savefig('images/observation_noise.png')
plt.show()
```
